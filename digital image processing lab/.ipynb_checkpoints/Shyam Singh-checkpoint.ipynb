```python
import numpy as np
import pandas as pd
import cv2 as cv      #Conda install -c forge openCV
import matplotlib.pyplot as plt
```

# Reading , writing and display the image

```python
# Read the image
image = cv.imread('bird.jpeg')

# Check if the image was successfully loaded
if image is not None:
#     print("Image loaded successfully.")
    # Display the image
    cv.imshow('Image', image)
    cv.waitKey(0)  # Wait for any key press to close the window
    cv.destroyAllWindows()  # Close all OpenCV windows
else:
    
    print("Error: Unable to load the image.")
```

# Separating different channels from the image

```python
# Assuming "image" is your original image in BGR format
# Convert BGR image to grayscale

img_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    
# Display the grayscale image
plt.imshow(img_gray)
```

```python
# Convert the image to grayscale
img_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    
# Display the grayscale image
cv.imshow('Grayscale Image', img_gray) #cv vaali ye line load nhi ho rhi h  , cv vaali line se ek alag popup aata h jisme image hoti h but plt vaali line se plot hota h image with graph
```
