```python
import cv2

# Assuming "image" is your original image in BGR format
# Convert BGR image to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,label='forest')
```

```python
image_bgr = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
plt.imshow(image_bgr,label='forest')
```

```python
image_rgb = cv2.cvtColor(image_gray, cv2.COLOR_GRAY2RGB)
plt.imshow(image_rgb,label='forest')
```

```python
plt.grid('got.jpg')
plt.xlabel('X_AXIS')
plt.ylabel('Y_AXIS')
plt.imshow(image)
```

```python

angle = np.radians(20)
x_img = np.float32([
    [np.cos(angle), -np.sin(angle), 0],
    [np.sin(angle), np.cos(angle), 0],
    [0, 0, 1]
])

rotated_img = cv.warpPerspective(image, x_img, (image.shape[1], image.shape[0]))

plt.imshow(rotated_img)
plt.show()
```
